# Basic Questions

**1. Which team won the most matches?**

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index, x = match_wins.values)
plt.title('Most Match wins by Team')
plt.xlabel('Count')
plt.ylabel('Teams')
```

**2. Toss Decision Trends**

```python
sns.countplot(x = df['toss_decision'])
plt.title('Toss Decision Trends')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
```

**3. Toss winner vs Match Winner**

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
```

```python
percentage.round(2)
```

**4. How do teams win? (Runs vs Wickets)**

```python
sns.countplot(x = df['won_by'])
plt.title('Won by')
plt.xlabel('Category')
plt.ylabel('Count')
```

# Key Player Performances

**1. Most "Player of the Match" Award**

```python
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x = count.values, y = count.index)
plt.title("Most Player of the Match")
plt.xlabel("Count")
plt.ylabel("Player Name")
```

**2. 2 Top Scorers**

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(2)

high.plot(kind = 'barh')
plt.title("Top 2 Scorer")
plt.xlabel("Score")
plt.ylabel("Player Name")
```

**3. 10 Best Bowling Figure**

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlors = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False).head(10)

top_bowlors.plot(kind = 'barh')
plt.title("Top 10 Bowlers")
plt.xlabel("Wickets")
plt.ylabel("Player Name")
```

# Venue Analysis

**Most Matches Played by Venue**

```python
venue_count = df['venue'].value_counts()
sns.barplot(y = venue_count.index, x = venue_count.values)
plt.title("Most Matches Played by Venue")
plt.xlabel("Count")
plt.ylabel("Venue")
```
